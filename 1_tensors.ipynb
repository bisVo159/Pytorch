```python
# prompt: create 2 tensor of 1000*10000 size and do matrix multiplication on cpu and gpu and calculate time

import time

size = (1000, 10000)

# CPU
tensor1_cpu = torch.rand(size)
tensor2_cpu = torch.rand(size).T  # Transpose for valid matrix multiplication

start_time_cpu = time.time()
result_cpu = torch.matmul(tensor1_cpu, tensor2_cpu)
end_time_cpu = time.time()
cpu_time = end_time_cpu - start_time_cpu

print(f"CPU matrix multiplication time: {cpu_time:.4f} seconds")

if torch.cuda.is_available():
    # GPU
    device = torch.device('cuda')
    tensor1_gpu = torch.rand(size, device=device)
    tensor2_gpu = torch.rand(size, device=device).T  # Transpose for valid matrix multiplication

    start_time_gpu = time.time()
    result_gpu = torch.matmul(tensor1_gpu, tensor2_gpu)
    torch.cuda.synchronize() # Wait for GPU to finish computation
    end_time_gpu = time.time()
    gpu_time = end_time_gpu - start_time_gpu

    print(f"GPU matrix multiplication time: {gpu_time:.4f} seconds")
    print(f"Speedup: {cpu_time/gpu_time:.4f}")
else:
    print("GPU not available for matrix multiplication.")
```
